```python
lm = 3
lc = 3
rm = 0
rc = 0
boat = "left"

print("Game Start")
print("Move all missionaries and cannibals to the right side\n")

while True:
    print("\nLeft --> Missionaries:", lm, "Cannibals:", lc)
    print("Right --> Missionaries:", rm, "Cannibals:", rc)
    print("Boat is on", boat, "side")

    # Winning condition
    if lm == 0 and lc == 0:
        print("\n🎉 You won! Everyone crossed safely.")
        break

    m = int(input("Enter number of missionaries: "))
    c = int(input("Enter number of cannibals: "))

    # Boat capacity check
    if m + c > 2 or m + c == 0:
        print("Invalid move! Boat can carry 1 or 2 people only.")
        continue

    if boat == "left":
        lm -= m
        lc -= c
        rm += m
        rc += c
        boat = "right"
    else:
        lm += m
        lc += c
        rm -= m
        rc -= c
        boat = "left"

    # Negative check
    if lm < 0 or lc < 0 or rm < 0 or rc < 0:
        print("Invalid move! Not enough people.")
        break

    # Cannibals eat missionaries condition
    if (lm > 0 and lc > lm) or (rm > 0 and rc > rm):
        print("Cannibals ate the missionaries! Game Over.")
        break
```
